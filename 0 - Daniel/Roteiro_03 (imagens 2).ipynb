```python
import cv2

image = cv2.imread('/content/noise.tif', cv2.IMREAD_GRAYSCALE)
```

```python
hist_image = cv2.calcHist([image], [0], None, [256], [0, 255])
```

```python
(T, image_binario) = cv2.threshold(image, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print('limiar: ', T)

```

```python
import numpy as np
```

```python
imagem_suavizada = cv2.GaussianBlur(image, (0, 0), 1.5)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,4))
axes[0].imshow(image, cmap='gray')
axes[1].imshow(image_binario, cmap='gray')
axes[2].imshow(imagem_suavizada, cmap='gray')

plt.show()
```

```python
(T, image_bin) = cv2.threshold(imagem_suavizada, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print('limiar: ', T)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,4))
axes[0].imshow(image, cmap='gray')
axes[1].imshow(imagem_suavizada, cmap='gray')
axes[2].imshow(image_bin, cmap='gray')

plt.show()
```

```python
imagem_suave = cv2.medianBlur(image, 5)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,4))
axes[0].imshow(image, cmap='gray')
axes[1].imshow(image_binario, cmap='gray')
axes[2].imshow(imagem_suave, cmap='gray')

plt.show()
```

```python
(T, image_bin2) = cv2.threshold(imagem_suave, 0, 255, cv2.THRESH_BINARY+cv2.THRESH_OTSU)
print('limiar: ', T)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(10,4))
axes[0].imshow(image, cmap='gray')
axes[1].imshow(imagem_suave, cmap='gray')
axes[2].imshow(image_bin2, cmap='gray')

plt.show()
```
